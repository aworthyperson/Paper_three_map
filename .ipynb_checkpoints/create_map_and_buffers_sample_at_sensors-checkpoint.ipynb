```python
data_fxn = '/Users/amandaworthy/Paper3_data_collection/USEFULDATA/energy_data_tmyx_and_location_jan22.csv'

df = pd.read_csv(data_fxn)
```

```python
mygeometry_array = gpd.points_from_xy(df['Longitude (E+/W-)'], df['Latitude (N+/S-)'])

gdf= gpd.GeoDataFrame(
    df, crs='EPSG:4326', geometry=mygeometry_array)

gdf
```

```python
# Convert to a suitable projected CRS for accurate buffering in meters
gdf = gdf.to_crs(epsg=3395)  

# Define the range of buffer distances
buffer_distances = range(10, 1000, 30)  # 50 to 5000 w 100 m intervals

# Create buffers for each distance and store them in separate columns
for distance in buffer_distances:
    gdf[f'{distance}m_buffer'] = gdf['geometry'].buffer(distance)

gdf
```

```python
import geopandas as gpd

# Assuming gdf is your existing GeoDataFrame with buffer columns and a `location_type` column

# Create a dictionary to store the new GeoDataFrames
buffered_gdfs = {}

# Iterate through the columns in the gdf that contain the word 'buffer'
for column in gdf.columns:
    if 'buffer' in column:
        # Extract the buffer size from the column name (assuming format like '50m_buffer')
        buffer_size = int(column.split('m')[0])  # Extract the numerical part before 'm_buffer'
        
        # Create a new GeoDataFrame with the buffer column as the geometry, include the `location_type` column, and add buffer size
        buffer_gdf = gpd.GeoDataFrame({
            'WMO': gdf['WMO'],
            'geometry': gdf[column],
            'buffer_size': buffer_size  # Add the buffer size as a new column
        }, crs=gdf.crs)
        
        # Store the new GeoDataFrame in the dictionary with the buffer distance as the key
        buffered_gdfs[column] = buffer_gdf

# Check the keys in the dictionary to confirm
print(buffered_gdfs.keys())

# Example to view the added buffer_size column
for key, df in buffered_gdfs.items():
    print(f"\nDataFrame for {key}:")
    print(df.head())
```

```python
# Concatenate all GeoDataFrames in buffered_gdfs into a single GeoDataFrame
concatenated_gdf = pd.concat(buffered_gdfs.values(), ignore_index=True)

# Check the result
concatenated_gdf
```

```python
# Find Length of Streets  

streets_fn = '/Users/amandaworthy/PHI_Big_Files/Seattle_Streets_2308916486970203731.geojson'

streets = gpd.read_file(streets_fn, crs='EPSG:4326')

streets = streets.to_crs('EPSG:3395')

streets.crs

streets.plot();
```

```python
test= concatenated_gdf.copy()
```

```python
import geopandas as gpd
from shapely.ops import unary_union
import time

# Start timing
start_time = time.time()

# Combine the streets into a union
combined_lines = streets.geometry.unary_union
streets_combined = gpd.GeoDataFrame(geometry=[combined_lines], crs=streets.crs)

# Calculate intersections in all rows
intersections = []
total_rows = len(test)  # Total number of rows in the test GeoDataFrame

# Iterate over each row in the `test` GeoDataFrame
for idx, row in test.iterrows():
    # Calculate the intersection of the buffer with the streets_combined
    intersection = streets_combined.intersection(row['geometry'])
    intersection = unary_union(intersection)
    intersections.append(intersection)
    
    # Calculate and print percentage completion
    percent_complete = (idx + 1) / total_rows * 100
    elapsed_time = time.time() - start_time
    print(f"Progress: {percent_complete:.2f}% complete. Elapsed time: {elapsed_time:.2f} seconds.")

# Create a GeoDataFrame with the results
result = gpd.GeoDataFrame({'geometry': intersections}, crs=test.crs)


# Add the intersections and lengths to the original `test` GeoDataFrame
test['streets_intersection'] = result['geometry']
test['streets_length'] = test['streets_intersection'].length


# Print the total time taken
total_time = time.time() - start_time
print(f"Total time taken: {total_time:.2f} seconds.")
```

## THE TREE CANOPY COVERAGE DATASET
-- DUH DUH DUH 

```python
from rasterio.crs import CRS
import rioxarray as rxr
import xarray as xr

tcc_path= '/Users/amandaworthy/Paper_3_GEE_files/tcc_bigger_seattle.tif'

dataarray_tcc= xr.open_rasterio(tcc_path)

dataarray_tcc
```

```python
import geopandas as gpd
import rioxarray
import numpy as np
import time

# Ensure the CRS of the GeoDataFrame matches the CRS of the raster
gdf= test.to_crs(dataarray_tcc.rio.crs)

# Function to calculate the average tree canopy coverage for each geometry
def calculate_average_canopy(geometry, raster):
    try:
        # Clip the raster using the geometry
        clipped_raster = raster.rio.clip([geometry], gdf.crs, drop=True)
        
        # Calculate the mean of the non-NaN values
        mean_canopy = clipped_raster.mean().item()
        
        # Handle cases where the mean is NaN (e.g., no intersection)
        if np.isnan(mean_canopy):
            return None  # or return 0 if you prefer to use 0 as a default value
        
        return mean_canopy
    
    except Exception as e:
        # Handle any unexpected errors
        print(f"Error processing geometry: {e}")
        return None

# Start timing
start_time = time.time()

# Total number of rows
total_rows = len(gdf)

# Create a list to store average canopy values
avg_canopy_values = []

# Iterate through each geometry and calculate the average canopy
for i, geometry in enumerate(gdf['geometry']):
    avg_canopy = calculate_average_canopy(geometry, dataarray_tcc)
    avg_canopy_values.append(avg_canopy)
    
    # Calculate and print progress every 5%
    if (i + 1) % (total_rows // 20) == 0:  # 20 intervals (5% each)
        elapsed_time = time.time() - start_time
        progress = (i + 1) / total_rows * 100
        print(f"Progress: {progress:.2f}% - Time elapsed: {elapsed_time:.2f} seconds")

# Assign the calculated values to the GeoDataFrame
gdf['avg_canopy'] = avg_canopy_values

# End timing
end_time = time.time()
total_elapsed_time = end_time - start_time

# Print the total elapsed time
print(f"Total elapsed time for calculating average canopy coverage: {total_elapsed_time:.2f} seconds")

# Display the first few rows of the GeoDataFrame
gdf.head()

```

```python
import matplotlib.pyplot as plt

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 10))


# Plot the GeoDataFrame with 'avg_canopy' as the color on top of the base map
gdf.plot(column='avg_canopy', cmap='Greens', legend=True, 
         legend_kwds={'label': "Average Tree Canopy Coverage"},
         ax=ax) 

# Plot the base map (Seattle GeoDataFrame)
seattle_gdf.plot(ax=ax, edgecolor="black", facecolor="none")



# Add a title and axis labels
plt.title("Tree Canopy Coverage by Geometry")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Display the plot
plt.show()
```

```python
from scipy import stats

# Ensure the CRS of the GeoDataFrame matches the CRS of the raster
gdflc = gdf.to_crs(landcover.rio.crs)

# Function to calculate the mode of land cover for each geometry
def calculate_mode_landcover(geometry, raster):
    try:
        # Clip the raster using the geometry
        clipped_raster = raster.rio.clip([geometry], gdflc.crs, drop=True)
        
        # Flatten the raster values to 1D array
        values = clipped_raster.values.flatten()
        
        # Remove NaNs
        values = values[~np.isnan(values)]
        
        if len(values) == 0:
            return None  # No data in the clipped raster
        
        # Calculate the mode of the values
        mode_value, count = stats.mode(values, axis=None)
        
        return mode_value[0]
    
    except Exception as e:
        # Handle any unexpected errors
        print(f"Error processing geometry: {e}")
        return None

# Start timing
start_time = time.time()

# Total number of rows
total_rows = len(gdflc)

# Create a list to store mode land cover values
mode_landcover_values = []

# Iterate through each geometry and calculate the mode land cover
for i, geometry in enumerate(gdflc['geometry']):
    mode_landcover = calculate_mode_landcover(geometry, landcover)
    mode_landcover_values.append(mode_landcover)
    
    # Calculate and print progress every 5%
    if (i + 1) % (total_rows // 20) == 0:  # 20 intervals (5% each)
        elapsed_time = time.time() - start_time
        progress = (i + 1) / total_rows * 100
        print(f"Progress: {progress:.2f}% - Time elapsed: {elapsed_time:.2f} seconds")

# Assign the calculated values to the GeoDataFrame
gdflc['mode_landcover'] = mode_landcover_values

# End timing
end_time = time.time()
total_elapsed_time = end_time - start_time

# Print the total elapsed time
print(f"Total elapsed time for calculating mode land cover: {total_elapsed_time:.2f} seconds")

# Display the first few rows of the GeoDataFrame
gdflc.head()
```

```python
import matplotlib.pyplot as plt

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 10))


# Plot the GeoDataFrame with 'avg_canopy' as the color on top of the base map
gdflc.plot(column='mode_landcover', cmap='tab20', legend=True, 
         legend_kwds={'label': "Landcover Mode"},
         ax=ax)  # Ensure this plot is added to the existing ax


# Plot the base map (Seattle GeoDataFrame)
seattle_gdf.plot(ax=ax, edgecolor="black", facecolor="none")


# Add a title and axis labels
plt.title("Landcover Mode by Geometry")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Display the plot
plt.show()
```
